```python
from matplotlib import pyplot as pit

x=[5,8,10]
y=[12,16,6]

pit.plot(x,y)
pit.title("Title")
pit.xlabel("Yaxis")
pit.ylabel("X axis")

pit.show()
```

```python
from matplotlib import pyplot as pit

x=[5,8,10]
y=[12,16,6]

x1=[6,9,10]
y1=[15,14,5]
pit.plot(x,y,'g',label='first',linewidth="6")
pit.plot(x1,y1,'b',label='second',linewidth=6)
pit.grid(True,color='r')
pit.legend()
pit.title("Title")
pit.xlabel("Yaxis")
pit.ylabel("X axis")

pit.show()
```

```python
from matplotlib import pyplot as pit

x=[5,8,10]
y=[12,16,6]
x1=[6,9,11]
y1=[6,15,7]


pit.bar(x,y,color='g',align="center")
pit.bar(x1,y1,color='r',align="center")

pit.title("Bar Graph")

pit.xlabel("X axis")
pit.ylabel("Y axis")

pit.show()
```

```python
from matplotlib import pyplot as pit

x=[5,8,10]
y=[12,16,6]
x1=[6,9,11]
y1=[6,15,7]


pit.scatter(x,y,color='g',label="first")
pit.scatter(x1,y1,color='r',label="second")

pit.title("Bar Graph")

pit.xlabel("X axis")
pit.ylabel("Y axis")
pit.legend()

pit.show()
```

```python
import matplotlib.pyplot as pit

labels=['Cookies','Jellbean','Milkshake','Cheeseeake']
sizes=[38.4,40.6,20.7,10.3]
colors=['yellowgreen','gold','red','lightcoral']
patches,texts=pit.pie(sizes,colors=colors,shadow=True,startangle=90)
pit.legend(patches,labels,loc='best')
pit.axis("equal")
pit.tight_layout()
pit._show()
```
